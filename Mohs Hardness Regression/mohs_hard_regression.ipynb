```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

#Train

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/data/mohs hard train.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
sns.pairplot(df, x_vars=['allelectrons_Total', 'density_Total', 'allelectrons_Average',
       'val_e_Average', 'atomicweight_Average', 'ionenergy_Average',
       'el_neg_chi_Average', 'R_vdw_element_Average', 'R_cov_element_Average',
       'zaratio_Average', 'density_Average'], y_vars='Hardness',height=7,aspect=0.7, kind='reg')
```

```python
cols = ['allelectrons_Total', 'density_Total', 'allelectrons_Average',
       'val_e_Average', 'atomicweight_Average', 'ionenergy_Average',
       'el_neg_chi_Average', 'R_vdw_element_Average', 'R_cov_element_Average',
       'zaratio_Average', 'density_Average']
```

```python
X = df[cols]
```

```python
y = df['Hardness']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
list(zip(cols,linreg.coef_))
```

```python
y_pred = linreg.predict(X_test)
```

```python
print(metrics.mean_absolute_error(y_test,y_pred))
```
