```python
def fill_missing_values(group):
    # Handle 'HomePlanet'
    if group['HomePlanet'].isnull().all():
        mode_value = df['HomePlanet'].mode().iloc[0]
        group['HomePlanet'] = group['HomePlanet'].fillna(mode_value)
    elif group['HomePlanet'].isnull().any():
        mode_value = group['HomePlanet'].mode()
        if not mode_value.empty:
            group['HomePlanet'] = group['HomePlanet'].fillna(mode_value.iloc[0])

    # Handle 'Destination'
    if group['Destination'].isnull().all():
        mode_value = df['Destination'].mode().iloc[0]
        group['Destination'] = group['Destination'].fillna(mode_value)
    elif group['Destination'].isnull().any():
        mode_value = group['Destination'].mode()
        if not mode_value.empty:
            group['Destination'] = group['Destination'].fillna(mode_value.iloc[0])

    # Handle 'Cabin'
 #   if group['Cabin'].isnull().all():
  #      mode_value = df['Cabin'].mode().iloc[0]
   #     group['Cabin'] = group['Cabin'].fillna(mode_value)
    #elif group['Cabin'].isnull().any():
     #   mode_value = group['Cabin'].mode()
      #  if not mode_value.empty:
       #     group['Cabin'] = group['Cabin'].fillna(mode_value.iloc[0])

 # Handle 'Deck'
    if group['Deck'].isnull().all():
        mode_value = df['Deck'].mode().iloc[0]
        group['Deck'] = group['Deck'].fillna(mode_value)
    elif group['Deck'].isnull().any():
        mode_value = group['Deck'].mode()
        if not mode_value.empty:
            group['Deck'] = group['Deck'].fillna(mode_value.iloc[0])

 # Handle 'Side'
    if group['Side'].isnull().all():
        mode_value = df['Side'].mode().iloc[0]
        group['Side'] = group['Side'].fillna(mode_value)
    elif group['Side'].isnull().any():
        mode_value = group['Side'].mode()
        if not mode_value.empty:
            group['Side'] = group['Side'].fillna(mode_value.iloc[0])

 # Handle 'CabNum'
    if group['CabNum'].isnull().all():
        mode_value = df['CabNum'].mode().iloc[0]
        group['CabNum'] = group['CabNum'].fillna(mode_value)
    elif group['CabNum'].isnull().any():
        mode_value = group['CabNum'].mode()
        if not mode_value.empty:
            group['CabNum'] = group['CabNum'].fillna(mode_value.iloc[0])

    # Handle 'Lname'
    if group['Lname'].isnull().all():
        mode_value = df['Lname'].mode().iloc[0]
        group['Lname'] = group['Lname'].fillna(mode_value)
    elif group['Lname'].isnull().any():
        mode_value = group['Lname'].mode()
        if not mode_value.empty:
            group['Lname'] = group['Lname'].fillna(mode_value.iloc[0])

    return group

# Apply the function on the dataframe grouped by 'GroupNum'
df1 = df.groupby('GroupNum').apply(fill_missing_values)
```
